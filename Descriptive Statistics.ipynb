```python
cov = {}
for col in data.columns:
    cov[col] = {}
    for other in data.columns:
        if col != other:
           covariance = data[col].cov(data[other])
           cov[col][other] = covariance
```

```python
print("covariance from each column")
cov_df = pd.DataFrame.from_dict(cov)
cov_df
```

how covariance works


```python
# Calculate the average of each column
mean_sugar = data["residual sugar"].mean()
mean_sulfur = data["total sulfur dioxide"].mean()

sum_product_diff = 0

# Iterate over each row to calculate the covariance components
for index, row in data.iterrows():
    sugar_diff = row["residual sugar"] - mean_sugar
    sulfur_diff = row["total sulfur dioxide"] - mean_sulfur
    sum_product_diff += sugar_diff * sulfur_diff

# Calculate covariance
covariance_sugar_sulfur = sum_product_diff / (len(data) - 1)

print("Covariance between residual sugar and total sulfur dioxide:", round(covariance_sugar_sulfur, 2))
```

```python
#make probability dictionary
probability = {}
for value in data["quality"]:
    if value in probability:
        probability[value] += 1
    else:
        probability[value] = 1

# Change to probability
total = len(data)
for key in probability:
    probability[key] = round(probability[key] / total, 4)
probability
```

```python
#convert to dataframe
probability_df = pd.DataFrame(list(probability.items()), columns=["Quality", "Probability"])
probability_df
```

```python
probability_df["xp"] = probability_df["Quality"] * probability_df["Probability"]
probability_df
```

Quality most spread out from >= 5 , therefore we must upgrade those which have quality < 5

Expected Value ----

```python
round(float(probability_df["xp"].sum()), 2)
```

On average, a randomly selected wine from this dataset will have a quality of around 5.6367.

not much different from actual mean

The difference between the expected value and the actual mean indicates that the probability distribution used does not fully represent the quality distribution. This could be due to the use of a subset of the data, uneven data distribution, or an inaccurate probability calculation method.

# Conclusion

Increasing levels of alcohol, sulfates, and citric acid, which have been statistically shown to be positively correlated with wine quality, significantly impacted the final result. When these three characteristics were above the average found in high-quality wines, 72.34% of the wines were classified as high quality. These findings suggest that these three characteristics can be key benchmarks in production strategies to consistently improve wine quality.

These findings have important implications for strategic decision-making. By understanding that the combination of three specific chemical features significantly increases the likelihood of achieving high quality, manufacturers can focus more on controlling and improving these aspects of the production process
